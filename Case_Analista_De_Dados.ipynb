#Agencias


```python
import pandas as pd
import numpy as np

agencias = pd.read_excel('202106AGENCIAS.xlsx')
agencias.columns = ['municipio', 'uf']
agencias.head(30)
```

```python

num_agencias_cidade = agencias['municipio'].value_counts().to_frame()
num_agencias_cidade

```

#Pontos


```python
postos = pd.read_excel('202106POSTOS.xlsx')
postos.columns = ['municipio', 'uf']
postos.head(30)
```

```python
num_postos_cidade = postos['municipio'].value_counts().to_frame()
num_postos_cidade
```

#PAE

```python
pae = pd.read_excel('202106PAE.xlsx')
pae.columns = ['municipio', 'uf']
pae.head(30)
```

```python
num_pae_cidade = pae['municipio'].value_counts().to_frame()
num_pae_cidade
```

#População

```python
populacao = pd.read_excel('CIDADE_POPULAÇÃO.xlsx')
populacao.columns = ['municipio', 'pop']
populacao
```

#Calculo


```python
concatenado = pd.concat([agencias, postos, pae])
b = concatenado.drop_duplicates()
b
tabela = pd.DataFrame()
tabela = concatenado['municipio'].value_counts().rename_axis('unique_values').to_frame('n_agencias')
df_value_counts = tabela.reset_index()
df_value_counts.columns = ['municipio', 'n_agencias']
df_value_counts = pd.merge(df_value_counts, b)
df_value_counts
```

#Nivel de Bancarização

```python
nb = pd.merge(df_value_counts, populacao)
nb['nivel_bancarizacao'] = nb['n_agencias'] / nb['pop']
nb.head()
```

Cidades mais desbancarizadas do Brasil

```python
del(nb['n_agencias'])
```

#Entrega Final


Tabela com Cidade com menor nivel de bancarização:

```python
nb = nb.sort_values(by=['nivel_bancarizacao'])
nb.columns = ['Nome da Cidade', 'Estado', 'Populacao', 'Nivel de Bancarizacao']

nb.head(60)
```
